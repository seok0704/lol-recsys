### Scrape All Summoners Above Platinum

```python
start_page = 1
plat_above = True
summoners = []

while plat_above:
    url = urllib.request.urlopen('https://na.op.gg/ranking/ladder/page='+str(start_page))
    content = url.read()
    soup = BeautifulSoup(content, 'lxml')
    
    start_page +=1
    
    containers = soup.findAll("tr", {"class": "ranking-table__row"})
    
    for container in containers:
        
        #Find the tier of the summoner
        summoner_tier = container.find("td",{"class":"ranking-table__cell ranking-table__cell--tier"}).get_text(strip=True)
        
        #Store basic summoner profile if they are gold+
        if "platinum 3" not in summoner_tier.lower():

            try:
                summoner_name = container.select_one("span").get_text()                
                summoner_lvl = container.find("td",{"class":"ranking-table__cell ranking-table__cell--level"}).get_text(strip=True)
                win = container.find("div",{"class":"winratio-graph__text winratio-graph__text--left"}).get_text(strip=True)
                loss = container.find("div",{"class":"winratio-graph__text winratio-graph__text--right"}).get_text(strip=True)
                
                summoners.append((summoner_name,summoner_tier, summoner_lvl, win, loss))

            except AttributeError:
                continue
        else:
            plat_above = False
            break
```

```python
# Create a table to summarize results
summonerDF = pd.DataFrame(summoners, columns=['Summoner','Tier','Summoner Level', 'Win','Loss'])
summonerDF.head()
```

```python
summonerDF['Total Match'] = summonerDF['Win'].astype(int) + summonerDF['Loss'].astype(int)
```

```python
export_csv = summonerDF.to_csv(r'summoner_list.csv')
```

Before scraping the player champion info, we need to get the list of champions in the game. The list is generated from the Riot API, Data Dragon where it stores the static information.

In production, when a user requests a recommendation for a specific summoner name, the model will request data from Riot API which returns back with the champion unique ID rather than the actual name of the champion. Therefore, we need the list of champions and its associated ID.

```python
champion_IDs = []
champion_names = []
champion_tags = []

for data in champion['data']:
    champion_IDs.append(data['key'])
    champion_names.append(data['name'])
    champion_tags.append(data['tags'])
    
champion_list = pd.DataFrame({'Champion ID': champion_IDs,
                             'Tags': champion_tags}, index = champion_names)
```

### Data Collection - Collect Played Champion List

Finally, using the list of summoners from the leaderboards, we will scrape individuals profile and get the number of matches played for each champions. The matches played for each champion will be the core data of our recommended system. 

```python
#Check if the html tag for number of wins/losses are empty/non-existent.
def check_empty(result):
    if not result:
        result = 0
    else:
        result = result.get_text(strip=True)
        result = re.sub('[^0-9]+','',result)
    
    return result
```

```python
player_champions.index
```

```python
champion_played_list = []

for summoner in tqdm(player_champions.index):
    
    #Slugify the summoner name for urls
    summoner_url_encode = urllib.parse.quote(summoner)
    
    #Get Dynamic URL
    url = urllib.request.urlopen('https://na.op.gg/summoner/champions/userName='+str(summoner_url_encode))
    content = url.read()
    soup = BeautifulSoup(content, 'lxml')
    
    #Get body tag for season 10 
    containers = soup.find("div",{"class":"tabItem season-15"})
    containers = soup.find("tbody",{"class":"Body"})
    
    if containers:
        #Get rows of each champion info
        champions = containers.findAll("tr",{"class":"Row"})

    #Get matches played for each champion and store them in the dataframe
        for champion in champions:

            # Get Champion name and match the format with Riot API
            champion_name = champion.find("td",{"class":"ChampionName Cell"}).get_text(strip=True)
            champion_ID = champion_list.loc[champion_name]['Champion ID']

            #
            win = check_empty(champion.find("div",{"class":"Text Left"}))
            loss = check_empty(champion.find("div",{"class":"Text Right"}))

            matches_played = int(win) + int(loss)
            
            normalized_matches_played = matches_played/int(summonerDF[summonerDF['Summoner']==summoner]['Total Match'])

            champion_played_list.append((summoner,champion_ID,champion_name,matches_played,normalized_matches_played))
    else:
        pass
```
